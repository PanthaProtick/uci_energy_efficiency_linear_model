```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
df=pd.read_csv('/content/ENB2012_data1.csv')
df.info()
```

```python
x=df.iloc[:,:-2].values
y1=df.iloc[:,-2].values
y2=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test=train_test_split(x,y1,test_size=0.2,random_state=1)
x2_train,x2_test,y2_train,y2_test=train_test_split(x,y2,test_size=0.2,random_state=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_regressor1=DecisionTreeRegressor(random_state=0)
tree_regressor2=DecisionTreeRegressor(random_state=0)
tree_regressor1.fit(x1_train,y1_train)
tree_regressor2.fit(x2_train,y2_train)
```

```python
tree_pred1=tree_regressor1.predict(x1_test)
tree_pred2=tree_regressor2.predict(x2_test)
```

```python
r2_score(y1_test,tree_pred1)
```

```python
r2_score(y2_test,tree_pred2)
```
